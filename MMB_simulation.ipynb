```python
class arm:
  def __init__(self,low,high,seed=0):
    self.mu_high=high
    self.mu_low=low
    self.pulls=[]
    self.rng=torch.Generator()
    if (seed!=0):
      self.rng.manual_seed(seed)
  def pull(self,effort):
    if(effort):
      self.pulls.append(torch.rand(1,generator=self.rng)<self.mu_high)
    else:
      self.pulls.append(torch.rand(1,generator=self.rng)<self.mu_low)
    return self.pulls[-1]
```
