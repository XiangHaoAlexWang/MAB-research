```python
import numpy as np
from typing import Literal
from numpy.typing import ArrayLike
```

```python
class NoisyEffortBanditEnv:
    """
    A noisy effort bandit enviorment.

    Parameters
    ----------
    mu_high: ArrayLike
        A list of the high means for each arm
    mu_low: ArrayLike
        A list of the low means for each arm
    effort_costs: ArrayLike
        A list of the effort costs for each arm
    seed:
        The seed to use for the random number generator
    
    Attributes
    ----------
    K: int
        The number of arms
    rng: Generator: the random number generator used.
    """
    def __init__(self, mu_high: ArrayLike, mu_low: ArrayLike, effort_costs: ArrayLike, seed=None):
        self.mu_high = np.array(mu_high, dtype=float)
        self.mu_low = np.array(mu_low, dtype=float)
        self.effort_costs = np.array(effort_costs, dtype=float)
        self.K = len(self.mu_high)
        self.rng = np.random.default_rng(seed) if seed != None else np.random.default_rng()

        assert self.mu_high.shape == self.mu_low.shape
        assert self.mu_high.shape == self.effort_costs.shape
        assert np.all(self.mu_high >= self.mu_low)

    def pull(self, arm: int, effort: Literal['H'] | Literal['L']):
        """
        Parameters
        ----------
        arm : int
            Arm index, from 0 to K-1.
        effort : str
            Either "H" for high effort or "L" for low effort.

        Returns
        -------
        reward : int
            Bernoulli reward, either 0 or 1.
        cost : float
            Effort cost paid by the arm.
        """
        # TODO for tomorrow:
        # if effort == "H", use mu_high[arm] and cost effort_costs[arm]
        # if effort == "L", use mu_low[arm] and cost 0
        assert effort in ['H', 'L']
        if effort == 'H':
            return (1 if self.rng.random() < self.mu_high[arm] else 0), self.effort_costs[arm]
        if effort == 'L':
            return (1 if self.rng.random() < self.mu_low[arm] else 0), 0
```

```python
testenv = NoisyEffortBanditEnv((0.8, 0.6), (0.7, 0.6), (0.05, 0))
for arm in range(2):
    for effort in 'H', 'L':
        pulls = []
        for _ in range(1000):
            pulls.append(testenv.pull(arm, effort)[0])
        print(f"Arm: {arm} Effort: {effort} Mean: {np.mean(pulls)}")
```
